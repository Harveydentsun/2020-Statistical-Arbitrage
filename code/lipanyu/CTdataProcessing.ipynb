```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.DataFrame()
for i in tqdm(range(174)):
    sub = pd.read_excel('DATA/CT_Data_Original.xlsx', usecols=np.arange(3*i, 3*i+3))
    col = sub.columns
    sub = sub.rename(columns={col[0]: 'Date', col[2]: col[1]+'_POS'})
    sub = sub[pd.notnull(sub.Date)]
    sub = sub.set_index('Date')
    if i == 0:
        df = sub
    else:
        df = pd.merge(df, sub, left_index=True, right_index=True, how='outer')
df.to_excel('CT_DATA.xlsx')
```
